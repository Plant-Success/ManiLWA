```python
import pandas as pd
import matplotlib.pyplot as plt
import umap.umap_ 
import umap.plot
import numpy as np 

import scipy

from sklearn.manifold import TSNE, Isomap
```

```python
df = pd.read_excel('../Data/Microarray expression, A. thaliana.xlsx',engine='openpyxl',index_col='AGI')
df.columns = [i.strip() for i in df.columns.values]
```

```python
bases = ['WT_intact','WT_decap','WT_decap_IAA','WT_NPA','max1','max2','max3']
for base in bases:
    slice_df = df[[f'{base}_{i}' for i in range(1,4)]]

    slice_df.to_csv(f'data/{base}.csv')
```

```python
import seaborn as sns
sns.set_theme()
sns.set_context('talk')
```

```python
slice_cols = ['max3_1', 'max3_2', 'max3_3'] # Example on MAX WT 

X = df[slice_cols].values

umap_embedding = umap.UMAP(random_state = 0).fit_transform(X)
tsne_embedding = TSNE().fit_transform(X)
isomap_embedding = Isomap().fit_transform(X)
```

```python
reducer = umap.UMAP(random_state = 0).fit(X)
```

```python

# UMAP
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10,10))
sns.scatterplot(x = umap_embedding[:,0], y = umap_embedding[:,1], ax = ax, color = 'C0')
ax.set_title('UMAP Embedding')
ax.set_xlabel('UMAP_0')
ax.set_ylabel('UMAP_1')

plt.show()
plt.close(fig)
# tSNE
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10,10))
sns.scatterplot(x = tsne_embedding[:,0], y = tsne_embedding[:,1], ax = ax, color = 'C1')
ax.set_title('tSNE Embedding')
ax.set_xlabel('tSNE_0')
ax.set_ylabel('tSNE_1')

plt.show()
plt.close(fig)
# ISOMAP
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10,10))
sns.scatterplot(x = isomap_embedding[:,0], y = isomap_embedding[:,1], ax = ax, color = 'C2')
ax.set_title('Isomap Embedding')
ax.set_xlabel('Isomap_0')
ax.set_ylabel('Isomap_1')

plt.show()
plt.close(fig)
```
